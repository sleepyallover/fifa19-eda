#### Group the players into a more general position:

Forwards/Striker  
Attacking Midfielder  
Central Midfielder  
Defensive Midfielder  
Side Midfielder  
Wing/Side Back  
Centre Back  
GK  

```python
def generalpos(pos):
    if pos=='RF'or pos=='LF' or pos=='CF' or pos=='RS'or pos=='LS'or pos=='ST':
        return 'Forward/Striker'
    elif pos=='LW'or pos=='RW':
        return 'Winger'
    elif pos=='RAM' or pos=='LAM'or pos=='CAM':
        return 'Attacking Midfielder'
    elif pos=='RWB' or pos=='LWB' or pos=='RB' or pos=='LB':
        return 'Wing/Side Back'
    elif pos=='LDM' or pos=='RDM' or pos=='CDM':
        return 'Defensive Midfielder'
    elif pos=='RCM' or pos=='LCM' or pos=='CM':
        return 'Central Midfielder'
    elif pos=='RCB' or pos=='LCB' or pos=='CB':
        return 'Centre Back'
    elif pos=='LM' or pos=='RM':
        return 'Side Midfielder'
    else:
        return pos
```

```python
df['General Position']=df['Position'].apply(lambda x:generalpos(x))
```

### Player Wages

```python
#cleaning up currencies
def numeric(cur):
    return cur[1:]
```

```python
df['Salary']=df['Wage'].apply(lambda x:numeric(x))
```

```python
df['Salary'].astype(str)
```

```python
def K(x):
    if 'K' in x:
        return x[:-1]+'000'
    else:
        return x
```

```python
df['Salary (K)']=(df['Salary'].apply(lambda x:K(x)).astype(int))/1000
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(y='Name', x='Salary (K)', data=df.sort_values(by='Salary (K)', ascending=False).head(10), edgecolor='black', palette='inferno')
plt.xlabel('Weekly Salary (K)')
plt.ylabel('Player')
plt.title('10 Highest Paid Footballers')
```
